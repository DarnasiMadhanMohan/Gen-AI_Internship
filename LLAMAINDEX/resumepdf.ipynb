```python
from fpdf import FPDF

# Helper function to create a more detailed resume with fake college names
def create_resume_with_college(name, job_title, skills, experience, education, college_name, file_name):
    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()
    
    pdf.set_font("Arial", size=16, style='B')
    pdf.cell(200, 10, txt=f"{name}", ln=True, align='C')
    pdf.ln(10)
    
    pdf.set_font("Arial", size=12, style='B')
    pdf.cell(200, 10, txt=f"Job Title: {job_title}", ln=True)
    pdf.ln(10)
    
    pdf.set_font("Arial", size=12)
    pdf.multi_cell(0, 10, txt="Skills:")
    pdf.set_font("Arial", size=12, style='I')
    pdf.multi_cell(0, 10, txt=skills)
    pdf.ln(10)
    
    pdf.set_font("Arial", size=12)
    pdf.multi_cell(0, 10, txt="Experience:")
    pdf.set_font("Arial", size=12, style='I')
    pdf.multi_cell(0, 10, txt=experience)
    pdf.ln(10)
    
    pdf.set_font("Arial", size=12)
    pdf.multi_cell(0, 10, txt="Education:")
    pdf.set_font("Arial", size=12, style='I')
    pdf.multi_cell(0, 10, txt=education)
    pdf.ln(10)

    # Adding a separate heading for the college name
    pdf.set_font("Arial", size=12, style='B')
    pdf.multi_cell(0, 10, txt="College Name:")
    pdf.set_font("Arial", size=12, style='I')
    pdf.multi_cell(0, 10, txt=college_name)
    pdf.ln(10)
    
    # Save to PDF file
    pdf.output(file_name)

# Sample data for different job roles, sorted from beginner to expert roles
jobs_data_with_college = [
    ("Chris Taylor", "IT Support Specialist", 
     "Windows OS, Mac OS, Networking, Hardware Troubleshooting, Help Desk", 
     "3+ years providing IT support in a corporate environment", 
     "A+ Certification, B.A. in Information Systems", 
     "Tech Valley University", 
     "it_support_specialist_resume.pdf"),
    
    ("Alice Johnson", "Graphic Designer", 
     "Photoshop, Illustrator, Branding, Typography, Print Design", 
     "Created branding and marketing collateral for multiple clients", 
     "B.F.A. in Graphic Design", 
     "Design Institute of Excellence", 
     "graphic_designer_resume.pdf"),
    
    ("Sara Wilson", "UX/UI Designer", 
     "Adobe XD, Figma, Wireframing, Prototyping, User Research", 
     "Experience designing user-friendly interfaces for web and mobile apps", 
     "B.A. in Graphic Design", 
     "Creative Arts University", 
     "ux_ui_designer_resume.pdf"),
    
    ("Jane Smith", "Software Developer", 
     "JavaScript, React, Node.js, Python, Git, Agile", 
     "Worked on full-stack web applications and RESTful API development", 
     "B.Sc. in Information Technology", 
     "CodeMasters College", 
     "software_developer_resume.pdf"),

    ("Michael Lee", "Project Manager", 
     "Project Management, Agile, Scrum, MS Project, Budget Management", 
     "Led multiple cross-functional projects to successful delivery", 
     "PMP Certification, B.A. in Business Administration", 
     "Global Business Academy", 
     "project_manager_resume.pdf"),
    
    ("David Brown", "Business Analyst", 
     "Data Analysis, Requirements Gathering, Process Mapping, SQL", 
     "Analyzed business processes and provided data-driven solutions", 
     "B.A. in Business Analytics", 
     "TechnoBusiness University", 
     "business_analyst_resume.pdf"),
    
    ("John Doe", "Data Scientist", 
     "Python, R, SQL, Machine Learning, Deep Learning, Data Visualization, Big Data", 
     "3+ years in data analysis, predictive modeling, and machine learning", 
     "B.Sc. in Computer Science, M.Sc. in Data Science", 
     "DataTech Institute of Research", 
     "data_scientist_resume.pdf"),
    
    ("Emily Davis", "Marketing Manager", 
     "SEO, SEM, Social Media, Content Strategy, Marketing Analytics", 
     "5+ years managing marketing teams and executing digital campaigns", 
     "B.A. in Marketing", 
     "Marketing Excellence University", 
     "marketing_manager_resume.pdf")
]

# Create PDF resumes for each job with fake college names
for job in jobs_data_with_college:
    create_resume_with_college(job[0], job[1], job[2], job[3], job[4], job[5], job[6])

print("PDFs generated successfully!")
```
