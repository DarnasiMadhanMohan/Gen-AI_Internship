```python
colors = {"Technology": "blue", "Furniture": "orange", "Office Supplies": "green"}

plt.figure(figsize=(10, 10))

# Scatter plot with bubble sizes based on Profit
scatter = plt.scatter(
    df["Sales"],
    df["Discount"],
    s=np.abs(df["Profit"]) / 10,  # Bubble size scaled
    c=df["Category"].map(colors),
    alpha=0.6,
    edgecolors="w",
    label=df["Category"].unique()
)

# Add legend for categories
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10)
           for category, color in colors.items()]
plt.legend(handles, colors.keys(), title="Category", loc="upper right")

# Add labels and title
plt.xlabel("Sales")
plt.ylabel("Discount")
plt.title("Sales vs Discount by Category")
plt.grid(True, linestyle="--", alpha=0.7)

plt.show()
```
