### **Visualization in Machine Learning**

* Visualization in machine learning is important for understanding the data, predicting the model's performance and predicting the insights of the data.

* Visualizing the raw data helps to understand its structure, patterns and relationships before applying any machine learning algorithm.

* The two libraries used in visualization are MatplotLib and seaborn libraries which are primarily describes the data but seaborn is easy to use where as MatplotLib require a bunch of instructions to perform visualization.

* This file deals with Basic examples of Visualization using both the libraries

***1. Read the file (csv)***

```python
import pandas as pd 
df = pd.read_excel("global_superstore/Global Superstore.xls")
```

* Columns plays an important role in visualization because they provide us the relationship between them and for other aspects also.


```python
df.columns 
```

***2. Import necessary libraries***

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

***3. Performing visualization***
* These are some examples of visualizations and these are used in analysis of dataset

**1. Line Plot**
* A line plot is a graphical representation where data points from the Sales column are connected by a continuous line, showing trends or variations over the DataFrame's index with labeled axes and a legend.

```python
plt.figure(figsize = (10,5))
plt.plot(df['Sales'], label = 'Sales')
plt.title('Line plot for Sales')
plt.xlabel('Index')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

**2. Scatter Plot**
* A graph where individual data points are plotted as dots, used to visualize the relationship or correlation between two variables.

```python
plt.figure(figsize = (10,5))
plt.scatter(df['Sales'], df['Profit'] , c= 'green', alpha = 0.5)
plt.title('Scatter Plot of Sales / Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.legend()
plt.show()
```

**3. Barplot**
* A bar plot represents given data with rectangular bars, where the length or height of each bar is proportional to the value it represents. It is commonly used to compare quantities across categories.

```python
plt.figure(figsize = (10,6))
plt.bar(df.index[:10], df['Sales'][:10], color = 'blue')
plt.title("Barplot plot of Sales")
plt.xlabel('Index')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

**4. Histplot( Histogram)**
* A histplot is a visualization that represents the distribution of a dataset by dividing it into bins and displaying the frequency of data points within each bin as rectangular bars.

```python
plt.figure(figsize = (10,5))
plt.hist(df['Discount'], bins = 10, color = 'purple', alpha = 0.7)
plt.title("Histplot on Disocunt")
plt.xlabel("Discount")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

**5. Boxplot**
* A boxplot (or box-and-whisker plot) is a statistical visualization that displays the distribution of a dataset through five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum, along with potential outliers.

```python
plt.figure(figsize = (10,5))
plt.boxplot([df['Sales'], df['Profit']], labels = ['Sales', 'Profit'])
plt.title('Boxplot of Sales and Profit')
plt.ylabel('Values')
plt.show()
```

**6. Pie-Chart**
* A pie chart is a circular statistical graphic divided into slices to illustrate the proportions or percentages of different categories within a dataset, where each slice's angle represents its relative size.
* Note: In this only 5 rows are taken.

```python
plt.figure(figsize = (7,7))
plt.pie(df['Quantity'][:5], labels = df.index[:5], autopct = '%1.1f%%')
plt.title('Pie chart Of Quantity')
plt.show()  
```

**7. Heatmap**
* A heatmap is a data visualization technique that uses a grid of colored cells to represent the magnitude of values in a matrix or dataset, with color intensity indicating the scale or frequency of the data.

```python
df1 = df[['Sales', 'Profit', 'Shipping Cost']]
df1.corr()
```

```python
plt.figure(figsize = (10,5))
df1 = df[['Sales','Profit','Shipping Cost']]
sns.heatmap(df1.corr(), cmap = 'YlGnBu', annot = True)
plt.title('Heatmap of Sales, Profit and Shipping Cost')
plt.show()
```

**8. Pairplot**
* A pairplot is a visualization in which pairwise relationships between numerical variables in a dataset are displayed through scatterplots, along with histograms or density plots for individual variable distributions on the diagonal.

```python
sns.pairplot(df)
plt.title('Pairplot of the dataset')
plt.show()
```

**9. KDE plot**
* A KDE (Kernel Density Estimate) plot is a smooth, continuous probability density curve that estimates the distribution of a dataset, offering a smoother alternative to histograms for visualizing the data’s frequency distribution.

```python
plt.figure(figsize = (10,5))
sns.kdeplot(df['Profit'], shade = True, color = 'Blue')
plt.title('KDE plot of Profit')
plt.xlabel('Profit')
plt.show()
```

**10. Violin Plot**
* A violin plot combines aspects of a boxplot and a KDE plot, displaying the distribution of a dataset through a symmetrical shape that shows the density of the data at different values, along with summary statistics like the median and quartiles.

```python
plt.figure(figsize = (10,6))
sns.violinplot(data = df[['Sales', 'Profit']])
plt.title('Violin Plot of Sales and Profit')
plt.ylabel('Values')
plt.show()
```

**11. Stacked Barplot**
* A stacked bar plot is a bar chart where multiple categories are stacked on top of each other within a single bar, allowing for the comparison of the total size across categories as well as the distribution of subcategories within each bar.

```python
plt.figure(figsize = (10,5))
x = df.index[:10]
plt.bar(x, df['Sales'][:10], label = 'Sales',  color = 'blue')
plt.bar(x, df['Profit'][:10], bottom = df['Sales'][:10], label = "Profit", color = 'green')
plt.title('Stacked Barplot of Sales and Profit')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

**12. Area Plot**
* An area plot is a variation of a line plot where the area beneath the line is filled with color, describing the magnitude of values over given categories, and showing cumulative totals or distributions.

```python
plt.figure(figsize = (10,5))
plt.fill_between(df.index[:50], df['Sales'][:50], color = 'blue', alpha = 0.5, label = 'Sales')
plt.fill_between(df.index[:50], df['Profit'][:50], color = 'green', alpha = 0.5, label = 'Profit')
plt.title('Filled Area plot of Sales and Profit')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
```

**13. Horizontal Bar Plot**
* A horizontal bar graph is a type of bar chart where the bars are displayed horizontally rather than vertically, used to compare different categories along the x-axis with the values represented by the length of the bars.

```python
plt.figure(figsize  = (10,5))
plt.barh(df.index[:10], df['Shipping Cost'][:10], color = 'Purple')
plt.title('Horizontal Barplot of Shipping Cost')
plt.xlabel('Shipping Cost')
plt.ylabel('Index')
plt.show()
```

**14. Multiple Line Plot**
*  A multiple line plot displays multiple lines on the same graph, each representing a different dataset or variable, allowing for the comparison of trends or relationships across different categories over the same x-axis.

```python
plt.figure(figsize = (10,5))
plt.plot(df['Sales'][:20], label = 'Sales', color = 'orange')
plt.plot(df['Profit'][:20], label = 'Profit', color = 'green')
plt.title('Line plot for Sales and Profit')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.show()
```

**15. Grouped Bar Plot**
* A grouped bar plot is a bar chart where bars for different categories are grouped together and displayed side-by-side within each category, allowing for the comparison of multiple subcategories across the main categories.

```python
import numpy as np
x = np.arange(10)
width = 0.4
plt.figure(figsize=(10, 5))
plt.bar(x - width/2, df['Sales'][:10], width, label='Sales', color='yellow')
plt.bar(x + width/2, df['Profit'][:10], width, label='Profit', color='pink')
plt.title('Grouped Bar Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.xticks(x, df.index[:10])
plt.legend()
plt.show()
```

**16. Stackplot**
* A stackplot is a visualization that displays multiple datasets stacked on top of each other, showing the cumulative value over time or categories. It is often used to visualize how different components contribute to the whole over a continuous variable (like time).

```python
plt.figure(figsize = (8,5))
plt.stackplot(range(len(df)), df['Sales'], df['Profit'], df['Shipping Cost'], labels = ['Sales', 'Profit', 'Shipping Cost'])
plt.title('Stackplot of Sales, Profit and Shipping Cost')
plt.legend()
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()
```

**17. Radar Chart**
* A radar chart (also known as a spider or web chart) is a graphical representation of multivariate data where each axis represents a different variable. Data points are plotted on each axis and connected to form a closed shape, making it useful for comparing multiple variables or entities.

```python
import math
def radar_chart(data, categories):
    N = len(categories)
    values = data + data[:1]
    angles = [n / float(N) * 2 * math.pi for n in range(N)]
    angles += angles[:1]

    ax = plt.subplot(111, polar=True)
    plt.xticks(angles[:-1], categories)

    ax.plot(angles, values, linewidth=2, linestyle='solid')
    ax.fill(angles, values, alpha=0.4)

categories = ["Sales", "Quantity", "Discount", "Profit", "Shipping Cost"]
data = [df["Sales"].mean(), df["Quantity"].mean(), df["Discount"].mean(), df["Profit"].mean(), df["Shipping Cost"].mean()]
plt.figure(figsize=(6, 6))
radar_chart(data, categories)
plt.title("Radar Chart Example")
plt.show()
```

**18. Annotated Plot**
* An annotated plot is a chart where additional information, such as text, arrows, or labels, is added to specific points or regions of the plot to provide context or highlight important features in the data.

```python
plt.figure(figsize=(10, 6))
plt.plot(df["Sales"], df["Profit"], marker='*')
for i, txt in enumerate(df["Profit"]):
    plt.annotate(txt, (df["Sales"][i], df["Profit"][i]))
plt.title("Annotated Plot Example")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

**19. 3D Scatter plot**
* A 3D scatter plot is a three-dimensional version of a scatter plot, where data points are plotted in a three-dimensional space, allowing for the visualization of relationships between three continuous variables.

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df["Sales"], df["Profit"], df["Quantity"], c='g', marker='o')
ax.set_xlabel("Sales")
ax.set_ylabel("Profit")
ax.set_zlabel("Quantity")
plt.title("3D Scatter Plot Example")
plt.show()
```

**20. Subplot**
* A subplot is a way to arrange multiple individual plots within a single figure or window, typically in a grid layout, enabling the comparison of different visualizations side by side. Each subplot is a separate plot but part of a larger overall figure.

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1
axs[0, 0].plot(df["Sales"], label="Sales")
axs[0, 0].set_title("Sales Line Plot")
axs[0, 0].legend()

# Plot 2
axs[0, 1].bar(range(len(df)), df["Profit"], color='orange', label="Profit")
axs[0, 1].set_title("Profit Bar Plot")
axs[0, 1].legend()

# Plot 3
axs[1, 0].scatter(df["Quantity"], df["Shipping Cost"], color='green', label="Shipping Cost vs Quantity")
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].legend()

# Plot 4
axs[1, 1].pie(df["Discount"], labels=range(len(df)), autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title("Discount Pie Chart")

plt.tight_layout()
plt.show()
```
