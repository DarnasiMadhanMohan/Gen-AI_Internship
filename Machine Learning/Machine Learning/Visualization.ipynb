```python
import math
def radar_chart(data, categories):
    N = len(categories)
    values = data + data[:1]
    angles = [n / float(N) * 2 * math.pi for n in range(N)]
    angles += angles[:1]

    ax = plt.subplot(111, polar=True)
    plt.xticks(angles[:-1], categories)

    ax.plot(angles, values, linewidth=2, linestyle='solid')
    ax.fill(angles, values, alpha=0.4)

categories = ["Sales", "Quantity", "Discount", "Profit", "Shipping Cost"]
data = [data["Sales"].mean(), data["Quantity"].mean(), data["Discount"].mean(), data["Profit"].mean(), data["Shipping Cost"].mean()]
plt.figure(figsize=(8, 8))
radar_chart(data, categories)
plt.title("Radar Chart Example")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace with your file path if necessary)
df = pd.read_excel("global_superstore/Global Superstore.xls")

# Pair Plot
sns.pairplot(df, vars=['Sales', 'Profit', 'Discount'])
plt.suptitle("Pair Plot: Sales & Profit, Profit & Discount", y=1.02)
plt.show()

# Pie Chart: Profit by Market
market_profit = df.groupby('Market')['Profit'].sum()
plt.figure(figsize=(8, 8))
plt.pie(market_profit, labels=market_profit.index, autopct='%1.1f%%', startangle=140)
plt.title("Profit by Different Markets")
plt.show()

# Pie Chart: Sales by Category
category_sales = df.groupby('Category')['Sales'].sum()
plt.figure(figsize=(8, 8))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=140)
plt.title("Sales by Different Categories")
plt.show()

# Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Sales', 'Profit', 'Discount', 'Shipping Cost', 'Quantity']])
plt.title("Box Plot: Sales, Profit, Discount, Shipping Cost, Quantity")
plt.xticks(rotation=45)
plt.show()

# Histogram: Sales
plt.figure(figsize=(10, 6))
plt.hist(df['Sales'], bins=30, color='blue', alpha=0.7)
plt.title("Histogram: Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()

# Histogram: Quantity
plt.figure(figsize=(10, 6))
plt.hist(df['Quantity'], bins=20, color='green', alpha=0.7)
plt.title("Histogram: Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()

# Bar Plot: Sales (First 50 Rows)
plt.figure(figsize=(12, 6))
plt.bar(range(50), df['Sales'][:50], color='purple', alpha=0.7)
plt.title("Bar Plot: Sales (First 50 Rows)")
plt.xlabel("Index")
plt.ylabel("Sales")
plt.show()

# Scatter Plot: Sales vs Discount
plt.figure(figsize=(10, 6))
plt.scatter(df['Sales'], df['Discount'], color='blue', alpha=0.6)
plt.title("Scatter Plot: Sales vs Discount")
plt.xlabel("Sales")
plt.ylabel("Discount")
plt.show()

# Line Plot: Quantity
plt.figure(figsize=(10, 6))
plt.plot(df['Quantity'], color='orange', linestyle='-' ) 
plt.title("Line Plot: Quantity")
plt.xlabel("Index")
plt.ylabel("Quantity")
plt.show()
```
