# remove html tags

```python
import re  #Regular Expressions
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

```python
df['reviewerName'] = df['reviewerName'].apply(lambda x: remove_html_tags(str(x)) if pd.notnull(x) else x)
df['reviewText'] = df['reviewText'].apply(lambda x: remove_html_tags(str(x)) if pd.notnull(x) else x)
# Uncomment if needed for 'summary'
df['summary'] = df['summary'].apply(lambda x: remove_html_tags(str(x)) if pd.notnull(x) else x)
```

# remove stopwords

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    if pd.isnull(text):  # Handle NaN or null values
        return text
    stop_words = set(stopwords.words('english'))
    # Ensure the input is a string and remove stopwords
    return ' '.join(word for word in str(text).split() if word.lower() not in stop_words)
```

```python
df['reviewerName'] = df['reviewerName'].apply(remove_stopwords)
df['reviewText'] = df['reviewText'].apply(remove_stopwords)
df['summary'] = df['summary'].apply(remove_stopwords)
```

# stemming

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "walk walks walking walked"
stem_words(sample)
```

```python
text1 = "Strong backgroung, good read, quite up to date. It takes a holistic approach to the subject. Lack of references is a bit surprising.Jorge"
stem_words(text1)
```

```python
from nltk.stem import SnowballStemmer
```

```python
snow = SnowballStemmer('english')
def stem_words(text):
    return " ".join([snow.stem(word) for word in text.split()])
```

```python
text1 = "Strong backgroung, good read, quite up to date. It takes a holistic approach to the subject. Lack of references is a bit surprising.Jorge"
stem_words(text1)
```

# lemmatizer

```python
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
# Download the WordNet corpus
nltk.download('wordnet')

# You might also need to download the 'omw-1.4' package for wordnet
nltk.download('omw-1.4')
```

```python
wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He aws running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations = "?:!.,;"
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)
    
sentence_words
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
    print("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos='v')))
```
